```python
# Check the number of nulls in each predictor
predictors = merge2.columns
container = {}
for col in predictors:
    """
    if df[col].isna().sum() < 1000 and col != 'transportation_issues':
        predictors.append(col)
    """
    container[str(col)] = merge2[col].isna().sum()

container_sorted = sorted(container.items(), key=operator.itemgetter(1))
print("Number of predictors:", len(container_sorted))
print("\nNull counts:")
for predictor in container_sorted:
    print(predictor[0], predictor[1])
```
